```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a data frame with x and y values
data = {'x': [1,2,3,4,5], 
        'y': [2,6,4,8,3]}
df = pd.DataFrame(data)

# Print the data frame
print(df)

# Plot the graph
plt.plot(df['x'], df['y'])
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Graph')
plt.show()
```

```python
from tkinter import *
root = Tk()
root.geometry("500x300")

Label(root, text="Python Registration Form", font="ariel 15 bold").grid(row=0, column=3)

namelabel = Label(root, text="Name")
Emaillabel = Label(root, text="Email")

namelabel.grid(row=1, column=1)
Emaillabel.grid(row=2, column=1)

genderlabel = Label(root, text='Gender')
male = Radiobutton(root, text='Male')
female = Radiobutton(root, text='Female')

genderlabel.grid(row=4, column=1)
male.grid(row=4, column=2, sticky="nsew")
female.grid(row=4, column=3, sticky="nsew")

SelectCountrylabel = Label(root, text="Select country")
YourCodingLanguagelabel = Label(root, text="Your Coding Language")

SelectCountrylabel.grid(row=5, column=1)
YourCodingLanguagelabel.grid(row=6, column=1)

namevalue = StringVar()
Emailvalue = StringVar()
YourCountryvalue = StringVar()
YourCodingLanguagevalue = StringVar()
checkvalue = IntVar()

nameentry = Entry(root, textvariable =namevalue)
Emailentry = Entry(root, textvariable =Emailvalue)
YourCountryentry = Entry(root, textvariable =YourCountryvalue)
YourCodingLanguageentry = Entry(root, textvariable =YourCodingLanguagevalue)

nameentry.grid(row=1, column=3)
Emailentry.grid(row=2, column=3)
YourCountryentry.grid(row=3, column=3)
YourCodingLanguageentry.grid(row=6, column=3)

checkbtn = Checkbutton(text="remember me?", variable = checkvalue)
checkbtn.grid(row=7, column=1)

def getvals():
    print("Submit button clicked")

Button(text="Submit", command=getvals).grid(row=7, column=3)

root.mainloop()
```
